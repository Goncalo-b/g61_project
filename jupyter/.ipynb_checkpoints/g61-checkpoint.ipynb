```python
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
```

```python
file = pd.read_csv("g61_Elearning_Certificates1.csv", sep=";", 
                   parse_dates=["issue_date", "online_date", "student_dob"], 
                   dayfirst=True).dropna(how="all")

file = file.dropna(axis=1, how="all")



```

```python
file.tail(20)
```

```python
file["platform_id"] = pd.to_numeric(file["platform_id"], errors="coerce").astype("Int64")
file["course_id"] = pd.to_numeric(file["course_id"], errors="coerce").astype("Int64")
file["certificate_id"] = pd.to_numeric(file["certificate_id"], errors="coerce").astype("Int64")
file["student_id"] = pd.to_numeric(file["student_id"], errors="coerce").astype("Int64")
int_columns = ["platform_id", "course_id", "certificate_id", "student_id"]
for col in int_columns:
    file[col] = pd.to_numeric(file[col], errors="coerce").astype("Int64")

# Garantir que online_date é datetime (caso não tenha sido detetado no read_csv)
file["online_date"] = pd.to_datetime(file["online_date"], errors="coerce").dt.date
file["issue_date"] = pd.to_datetime(file["online_date"], errors="coerce").dt.date
file["student_dob"] = pd.to_datetime(file["online_date"], errors="coerce").dt.date
```

```python
file
```

```python
# Classe: Student
df_student = file[['student_id', 'student_dob']].drop_duplicates(subset=['student_id'])

# Classe: Platform
df_platform = file[['platform_id']].drop_duplicates(subset=['platform_id'])

# Classe: Course (contém a online_date da classe Course)
df_course = file[['course_id', 'course_name', 'platform_id', 'online_date']].drop_duplicates(subset=['course_id'])

# Classe: Registre / Transaction (O histórico de notas)
df_registre = file[['student_id', 'course_id']]

# Classe: Certificate (O resultado final)
df_certificate = file[['certificate_id', 'student_id', 'course_id', 'issue_date']].dropna(subset=['certificate_id'])
```

```python
engine = create_engine('sqlite:///tabelafinal.db')

# Dicionário de tabelas para automação
classes_sql = {
    "students": df_student,
    "platforms": df_platform,
    "courses": df_course,
    "registre": df_registre,
    "certificates": df_certificate
}
```

```python
print("--- Resumo da Exportação ---")
for table_name, df_data in classes_sql.items():
    df_data.to_sql(table_name, con=engine, if_exists='replace', index=False)
    print(f"Tabela '{table_name}': {len(df_data)} registos processados.")

print("\nBase de dados 'elearning_system.db' criada com sucesso!")
```
